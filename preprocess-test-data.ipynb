```python
import pandas as pd
import locale
from datetime import datetime
from pandas.tseries.offsets import MonthEnd

def transform_dataframe(df: pd.DataFrame, calculation_type: str) -> pd.DataFrame:
    """Transform dataframes; "unpivot" the dataframe (melt) and add a column with datetime format."""
    
    locale.setlocale(locale.LC_TIME, "id-ID.utf8")
    
    melted_df = df.melt(id_vars="Bulan")
    melted_df.rename(columns={"variable": "Tahun", "value": calculation_type}, inplace=True)
    melted_df["Bulan-Tahun"] = melted_df["Bulan"] + "-" + melted_df["Tahun"].astype("str")
    melted_df["date"] = pd.to_datetime(melted_df["Bulan-Tahun"], format="%B-%Y")+MonthEnd(1)
    
    return melted_df.drop(columns=["Bulan-Tahun"])
```

```python
melted_inflationyoy_df = transform_dataframe(df=inflation_yoy_df, calculation_type="YoY")
melted_inflationyoy_df
```

```python
melted_inflationmtm_df = transform_dataframe(df=inflation_mtm_df, calculation_type="MtM")
melted_inflationmtm_df
```

```python
melted_inflationytd_df = transform_dataframe(df=inflation_ytd_df, calculation_type="Ytd")
melted_inflationytd_df
```

```python
joined_inlfation_df = melted_inflationyoy_df \
    .merge(melted_inflationmtm_df[["MtM", "date"]], how="left", on="date") \
    .merge(melted_inflationytd_df[["Ytd", "date"]], how="left", on="date")
```
